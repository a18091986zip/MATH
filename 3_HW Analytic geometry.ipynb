##Задание 3
напишите код, реализующий построение графиков
- окружности
- эллипса
- гиперболы

```python
def circle(R, x0, y0):
  X = [x*0.1 for x in range(0,101,1)]
  X_ = [-1*x*0.1 for x in range(0,101,1)]
  Y = [math.sqrt(R**2 - (x-x0)**2 )+y0 for x in X]
  Y_ = [-math.sqrt(R**2 - (x-x0)**2)+y0 for x in X]
  Y1 = [math.sqrt(R**2 - (x-x0)**2 )+y0 for x in X_]
  Y1_ = [-math.sqrt(R**2 - (x-x0)**2)+y0 for x in X_]
  plt.plot(X,Y)
  plt.plot(X,Y_)
  plt.plot(X_,Y1)
  plt.plot(X_,Y1_)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.axis('equal')
  plt.show()

circle(10,0,0)
```

```python
def elipse(a, b):
  X = [x*0.01 for x in range(0,301,1)]
  X_ = [-1*x*0.01 for x in range(0,301,1)]
  Y = [math.sqrt((1 - x**2/a**2)*b**2) for x in X]
  Y_ = [-math.sqrt((1 - x**2/a**2)*b**2) for x in X]
  Y1 = [math.sqrt((1 - x**2/a**2)*b**2) for x in X_]
  Y1_ = [-math.sqrt((1 - x**2/a**2)*b**2) for x in X_]
  
  plt.plot(X,Y)
  plt.plot(X,Y_)
  plt.plot(X_,Y1)
  plt.plot(X_,Y1_)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.axis('equal')
  plt.show()

elipse(3,1)
```

```python
def hiperbola(a, b):
  X = [x*0.01 for x in range(300,1000,1)]
  X_ = [-1*x*0.01 for x in range(300,1000,1)]
  Y = [math.sqrt(-(1 - x**2/a**2)*b**2) for x in X]
  Y_ = [-math.sqrt(-(1 - x**2/a**2)*b**2) for x in X]
  Y1 = [math.sqrt(-(1 - x**2/a**2)*b**2) for x in X_]
  Y1_ = [-math.sqrt(-(1 - x**2/a**2)*b**2) for x in X_]
  
  plt.plot(X,Y)
  plt.plot(X,Y_)
  plt.plot(X_,Y1)
  plt.plot(X_,Y1_)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.axis('equal')
  plt.show()

hiperbola(3,1)
```

```python
from scipy.optimize import fsolve

def equations(p):
  x,y = p
  return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1, y1 = fsolve(equations, (-1,2))
x2, y2 = fsolve(equations, (2,2))

# print(x1,y1)
# print(x2,y2)

print(f'Ответ: x in ({x1}, {x2})')
```
