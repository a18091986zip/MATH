$$\begin{equation*}
\begin{cases}
3x - 2y + 5z = 7\\
7x + 4y -8z = 3\\
5x -3y -4z = -12
\end{cases}
\end{equation*}$$

Сложим 1 и 3 уравнения и вычтем 2-е:

$$x = 9y - 9z -8 \,\,\,\,\,\,\,\,\,\,\,(1)$$

Сложим 1 и 2 уравнения и вычтем 3-е:

$$z = 22 - 5x - 5y = 22 - 5\cdot(9y-9z-8) - 5y = 45z - 50y + 62 \Rightarrow z = \frac{50y-62}{44} \,\,\,\,\,\,\,\,\,\,\, (2)$$

Подставляем z в 1:

$$x = \frac{-54y + 206}{44} \,\,\,\,\,\,\,\,\,\,\, (3)$$

Сложим все уравнения:

$$y = 15x - 7z + 2 \,\,\,\,\,\,\,\,\,\,\, (4)$$

Подставим (2) и (3) в (4):

$$y = \frac{-810y + 3090}{44} - \frac{350y - 434 }{44} + \frac{88}{44} \Rightarrow $$

$$x = 1 \,\,\,\,\,\, y = 3 \,\,\,\,\,\, z = 2$$


Каждое уравнение линейное, система линейная
