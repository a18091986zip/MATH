#Векторы

__Вектор__ - упорядоченная совокупность N вещественных чисел. 

\begin{equation*}
\overrightarrow X = \left[
\begin{array}{cccc}
x_{1}\\
x_{2}\\
...\\
x_{N}
\end{array}
\right]
\end{equation*}

$$\overrightarrow X = [x_1, x_2, ..., x_N]$$

__одномерным вектором__ называется направленный отрезок, который помимо числовой характеристики (длины) имеет и направление

__двумерный вектор__ - направленный отрезок, проекция которого на ось х ($x_1$)даст нам первую координату вектора, а на ось y ($y_1$) - вторую координату вектора

#Прямая на плоскости

__уравнение прямой на плоскости__

$y = kx + b$
k определяет угол наклона, b - сдвиг по оси х

$Ax + By + C = 0$ A и B одновременно не равны 0

$A(x-x_0) + B(y-y_0) + C = 0$ уравнение прямой, проходящей через точку $(x_0, y_0)$

$\frac{x-x_0}{l} = \frac{y-y_0}{m}$ каноническое уравнение прямой 


$\frac{x-x_0}{x_1 - x_0 } = \frac{y - y_0}{y_1 - y_0}$

две прямые __перпендикулярны__, если произведение их угловых коэффициентов равно -1, и не перпендикулярны, если оно не равно -1.

```python
x = np.linspace(-5,5,21)
y = 2*x
y2 = 2*x+1
y3 = 3*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#перпендикулярные прямые
x = np.linspace(-3,3,21)
y = 3*x+1
y2 = -1/3*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

#Линии и поверхности в пространстве

__уравнение плоскости в пространстве__

$$Ax + By + Cz + D = 0$$

вектор перпендикулярный данной плоскости задается тремя компонентами A,B,C

Для того, чтобы точка с координатами $x_0$,$y_0$,$z_0$ лежала на плоскости, её координаты должны удовлетворять уравнению:

$$A(x-x_0) + B(y-y_0) + C(z-z_0) = 0$$

Если какой-либо коэффициент плоскости = 0, то такое уравнение плокости называется неплным

- если D = 0, плоскость проходит через точку (0,0,0)
- если B = 0, плоскость становится параллельной оси y
- если A = 0, плоскость становится параллельной оси x
- если C = 0, плоскость становится параллельной оси z 

__уравнение плоскости в отрезках__

$$\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1$$


__уравнение прямой в пространстве__

как пересечение двух плоскостей в пространстве

$\frac{x-x_1}{x_2 - x_1 } = \frac{y - y_1}{y_2 - y_1} = \frac{z - z_1}{z_2 - z_1} $


__поверхности второго порядка__

$$a_{11}x^2 + 2a_{12}xy + a_{22}y^2 + a_{33}z^2 + 2a_{13}xz + 2a_{23}yz + 2a_{14}x + 2a_{24}y + 2a_{34}z + a_{44} = 0$$

__эллипсоид__

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1$$

__однополостный гиперболоид__

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1$$

__двухполостный гиперболоид__

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1$$

__конус второго порядка__

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 0$$

__элиптический параболоид__

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = z$$

__гиперболический параболоид__

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} = z$$

#Преобразования координат

__параллельный перенос__

$$X = x - a$$
$$Y = y - b$$

__параллельный перенос с поворотом системы координат__

$$X = (x-a)cos\alpha + (y-b)sin\alpha$$
$$Y = -(x-a)sin\alpha + (y-b)cos\alpha$$

__только поворот__

$$X = xcos\alpha + ysin\alpha$$
$$Y = -xsin\alpha + ycos\alpha$$

__линейное преобразование плоскости__

$$X = a_{11}x + a_{12}y + a_{13}$$
$$Y = a_{21}x + a{22}y + a_{23}$$

$\Delta = a_{11}a{22} -a_{12}a_{21}$ - преобразователь. Если $\Delta = 0$, то преобразование называется вырожденным

__афинные преобразования__

характерны тем, что переводят прямую в одной системе координат также в прямую в другой системе координат

параллельные прямые также остаются параллельными

определитель не равен нулю

среди афинных преобразований особую роль играет __ортогональное преобразование__ - такое линейное преобразование на плоскости, которое удовлетворяет соотношениям:

- $$a_{11}^2 + a_{21}^2 = 1$$ 
- $$a_{12}^2 + a_{22}^2 = 1$$ 
- $$a_{11}a_{12} + a_{21}a_{22} = 1$$ 

__полярные координаты__

задают точку расстоянием от начала координат и углом относительно некоторого заранее заданного направления

$$ x = Rcosα $$
 
$$ y = Rsinα$$

$$R = \sqrt{x^2 + y^2}$$

$$0\leq R<\infty$$
$$0\leq α<2\pi$$

#Графическое и численное решение уравнений

```python
from scipy.optimize import fsolve

def f(x):
  return (np.sin(x+1))

x0 = fsolve(f,5)
print(x0)

x = np.linspace(-3*np.pi, 3*np.pi,201)
plt.plot(x, f(x))
plt.plot([-10,10], [0,0], color = 'red', linewidth = 0.75, linestyle = '--')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2, 3,201)
plt.plot(x, (np.exp(x)-1)/x)
plt.plot(x, x**2-1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
  x,y = p
  return (y - x**2 + 1, np.exp(x) - x*y -1)

x1, y1 = fsolve(equations, (-2,-1))

print(x1,y1)
```
